````python
class Member:
    def __init__(self,name,age,email,address):
        self.name = name
        self.age = age
        self.email = email
        self.address = address
    def __str__(self):
        return "{},{},{},{}".format(self.name,
                                   self.age,
                                   self.email,
                                   self.address)
    def __eq__(self,other):
        if isinstance(other,Member):
            return self.__str__()==other.__str__()
        return True
    def as_dict(self): # 객체를 딕셔너리로 return
        return {'name':self.name,'age':self.age,'email':self.email,'address':self.address}
    def as_list(self): # 객체를 list로 return
        return [self.name,self.age,self.email,self.address]
````

````python
# 1. 딕셔너리 리스트를 json파일로 저장
# ensure_ascii 옵션의 역할 : ASCII 코드 범위를 넘어선 한글은 유니코드로 저장
# True(기본값) : ASCII코드 범위를 넘어선 한글은 유니코드로 저장
# False : 비ASCII 문자의 원래 형태로 저장
import json
with open('data/ch09_member1.json','w',encoding='utf-8')as jsonfile:
    json.dump(data,jsonfile,ensure_ascii=False,indent='')
````

````python
#data = [{'name': 'Hong','age': '20','email': 'kildong@hong.com','address': 'Seoul'},
````

````python
# 2. 객체리스트를 json파일로 저장(dump) 
class Member:
    def __init__(self,name,age,email,address):
        self.name=name
        self.age=age
        self.email=email
        self.address=address
    def as_dict(self):
        '객체를 딕셔너리로 바꿔 return(json.dump함수 사용에 필요)'        
        return{"name":self.name,"age":self.age,"email":self.email,"address":self.address}
    #return self.__dict__
    def __str__(self):
        return "{}님 {}살 {}, {}주거중".format(self.name,self.age,self.email,self.address)
    def __eq__(self,other):
        return True
    '''
    m1.__eq__(m2)호출시 두객채의 속성변수값이 모두 일치하는지 여부
    m1==m2 구현시 m1.__eq__(m2)호출
    '''
    if isinstance(other,Member):
        return self.__str__() == other.__str__()
    else:
        return False
````

## Customer 클래스 작성

        * Customer("홍길동", "010-9999-9999", "k@hong.com", 30, 5, "vvip")
        *__init__,__str__,as_dict(csv파일), to_txt_style(객체를 text스타일로 출력하기위해)

````python
class Customer:
    def __init__(self,name,phone,email,age,grade,etc):
        self.name = name
        self.phone = phone
        self.email = email
        self.age = age
        self.grade = grade
        self.etc = etc
    def as_dict(self):
        return self.__dict__
    def __str__(self):# **** 홍길동 010-9999-9999  k@hong.com  30  vvip
        return"{:>5}\t{:3}\t{:15}\t{:15}\t{:3}\t{}".format(self.grade*'*',
                                                        self.name,
                                                        self.phone,
                                                        self.email,
                                                        self.age,
                                                        self.etc)
    def to_txt_style(self):# **** 홍길동 010-9999-9999  k@hong.com  30  vvip
        return "{}, {}, {}, {}, {}, {}".format(
                                          self.name,
                                          self.phone,
                                          self.email,
                                          self.age,
                                          self.grade,
                                          self.etc)
````

````python
c1=Customer("홍길동","010-9999-9999","h@hong.com",30,4,"기타")
print(c1.as_dict())
print(c1)
````

````python
def to_customers(row): # "홍길동, 010-9999-9999,  k@hong.com,  30, 5, vvip\n"=>Customerabs객체
    data = row.strip().split(', ')
    name = data[0]
    phone = data[1]
    email = data[2]
    age = int(data[3])
    grade = int(data[4])
    etc = data[5]
    return Customer(name,phone,email,age,grade,etc)
````
