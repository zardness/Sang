```python
정규표현식
'''
   1. 참조:https://ko.wikipedia.org/wiki/%EC%A0%95%EA%B7%9C_%ED%91%9C%ED%98%84%EC%8B%9D
   2. 간략문법
       \d :(숫자와 매치 [0-9]와동일)
       \D : (숫자가 아닌 문자)
       \w : (영문자나 숫자.[a-zA-Z0-9]와 동일)
       \W : (영문자나 숫자가 아닌 문자)
       . : (문자)
       \.: (.)
       \s : ()
       {n,} : n회 반복
       {2,4} : 2-4회 반복 
       ? : (0번이나 1번의 의미)
       + : (1번 이상 반복)
       * : (0번이상 반복)
    3.정규표현식 연습장 : 
       ex1.전화번호 패턴:0\d{1,2}[\-)\s\.]\d{3,4}[\-\s\.]\d{4}
       ex2.이메일 패턴 : \w[a-z0-9_]+@\w+(\.\w+){1,2}
'''
```

data ="이름:홍길동, 주소:신림동, 전화:02-888-8888,\
핸드폰:010-7777-7777,이메일:abc@naver.com,ksd@gmail.com,bbc@lll.com\
특징:서해번쩍 동해번쩍,혈액형:B ㅋㅋ"
print(data)

```python
import re
tel_pattern = r"0\d{1,2}[\-)\s\.]\d{3,4}[\-\s\.]\d{4}"
tels = re.findall(tel_pattern, data) # 해당 패턴을 다 찾아줘. 없으면 빈 list([])
tels
```

```python
# 정규표현식 패턴과 처음 match되는것을 찾아줘
firsttel = re.search(tel_pattern, data)
firsttel
```

```python
phone_pattern = r"(\d{2,3})-(\d{3,4})-(\d{4})"
phone = re.search(phone_pattern, data)
print(phone.group(0))
print(phone.group(1))
print(phone.group(2))
print(phone.group(3))
```

```python
email_pattern = r"\w[a-z0-9_]+@\w+(?:\.(\.\w+){1,2}"
email = re.search(email_pattern, data)
```
