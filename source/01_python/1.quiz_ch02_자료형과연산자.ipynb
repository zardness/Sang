```python
#1. 이름과 나이 변수를 다음 형식으로 출력하도록 format() 함수를 이용해 형식화하세요
#[출력형식 : 홍길동님의 나이는 23 살입니다

name = '홍길동'
age = 23
print('{}님의 나이는 {}살입니다'.format(name,age))
print('{0}님의 나이는 {1}살입니다'.format(name,age))
print('%s님의 나이는 %d살입니다.'%(name,age))
print(f'{name}님의 나이는 {age}살입니다')
```

```python
# 2.두 정수를 입력받아 두 수의 덧셈 , 뺄셈 , 곱셈 , 나눗셈 , 몫 , 나머지를 출력하세요
a = int(input('첫번재 수'))
b = int(input('두번째 수'))
print(a+b) #print('num1 + num2 = ', (num1+num2))
print(a-b) #print('num1 - num2 = ', (num1-num2))
print(a*b) #print('num1 * num2 = ', (num1*num2))
print(a/b) #print('num1 / num2 = ', (num1/num2))
print(a//b) #print('num1 // num2 = ', (num1//num2))
print(a%b) #print('num1 % num2 = ', (num1%num2))
```

```python
# 3.문자열의 분리하기와 합치기 기능을 이용하여 ‘Hello 가 할당된 변수를 이용하여
#‘World 로 출력하세요
text = "Hello world"
print(text[-5],text[:5])
splited_text = text.split()
print(splited_text[1],splited_text[0])
```

```python
# 3.문자열의 분리하기와 합치기 기능을 이용하여 ‘Hello World’가 
    #할당된 변수를 이용하여 ‘World Hello’로 출력하세요
greeting = 'Hello world'
print(greeting[-5:],greeting[:5]) # 메일 받은 내용
splited_greeting = greeting.split()
print(splited_greeting[1], splited_greeting[0])
```

```python
# 4.x = ' abcdef ＇를 이용하여 bcdefa 로 출력하세요 문자 슬라이싱이용
x = 'abcdef'
print(x[1:]+x[0])
print(x[1:],x[0],sep='')
```

```python
# 5. x = ‘ abcdef 를 이용하여 ‘EDCBA 로 출력하세요
x = 'abcdef'
print(x.upper()[::-1])
print(x[::-1].upper())
```

```python
# 6.오늘의 온도를 섭씨온도로 입력받아 화씨 온도로 변환하는 프로그램을 작성하세요 .
c = int(input('섭씨온도를 입력하세요'))
f = (c*1.8)+32
if not c%5: # 메일 받은 내용
    print("섭씨 {}도는 화씨온도로 바꾸면 {:.0f}도 입니다".format(c, f))
else:
    print("섭씨 {}도는 화씨온도로 바꾸면 {:.2f}도 입니다".format(c, f))
```
