•파일이 없거나 쓰기 금지 로 인한 오류

# 2절. try~except로 예외처리
## 2.1 try~except
```
try:
    예외가 발생할가능성이 있는 명령어들 
except:
    예외가 발생했을경우 실행할 명령어들
```

````python
# 100을 입력받은 정수값으로 나눠 출력
x = int(input('정수를 입력하세요>'))
print('입력한 정수는{}'.format(x))
print('100을 입력한 정수로 나누면{:.2f}'.format(100/x))
#ZeroDivisionError: "division by zero"
#                   0을 입력했을때 예외타입
#ValueError:invalid literal for int() with base 10:"a"
#            invalid literal for int() with base 10:"a"
````

````python
# 100을 입력받은 정수값으로 나눠 출력
#ZeroDivisionError: "division by zero"
#ValueError:invalid literal for int() 
try : 
    x = int(input('정수를 입력하세요>'))
    print('입력한 정수는{}'.format(x))
    print('100을 입력한 정수로 나누면{:.2f}'.format(100/x))
except:
    print('나누기에 유효한 정수가 아닙니다.')
print('DONE')
````

````python
# 100을 나눌 유효한 정수를 입력할 때까지 정수를 입력받아 유효한 정수를 입력하면 입력한 정수와 100으로 나눈 수를 출력
````

````python
while True:
    try : 
        x = int(input('정수를 입력하세요>'))
        print('입력한 정수는{}'.format(x))
        print('100을 입력한 정수로 나누면{:.2f}'.format(100/x))
        break
    except:
        print('나누기에 유효한 정수가 아닙니다.')
    
print('DONE')
````
