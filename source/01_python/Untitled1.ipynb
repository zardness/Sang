# 1절. 리스트

## 1.1 리스트 만들기


```python
numbers = [1,3,5,9,10,12,13,18]
print(numbers[::-1])
print(numbers[2::2])
print(numbers[::])
```

```python
lst=[item for item in numbers if item%2==0]
[n for n in numbers if n>=10]
```

```python
lst = [item for item in numbers if item%2]
lst
```

```python
fruit = ['orange-red','apple-red','mango','kiwi']
[fruit for fruit in fruit if (len(fruit)>=5) & fruit.endswith('red')]
```

```python
[len(fruit) for fruit in fruit]
```
