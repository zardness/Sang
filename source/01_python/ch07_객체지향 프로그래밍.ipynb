# 2절. 생성자( `__init__()`)와 소멸자(`__del__()`)
```
생성자:__init(self,매개변수,..)객체가 생성될 때 자동 실행
      생성시 필요한 코드(인스턴스 변수 초기화)를 포함 
      없으면 default기본 생성자 자동 생성
소멸자:__del__(self) 객체가 소멸될(참조한 변수가 없어질)때 자동실행
      인스턴스 객체의 레퍼런스 카운트가 0이될때 실행
      소멸시 필요한 코드를 포함
```

````python
class Person:
    def __init__(self,name,gender):
        self.name = name   #인스턴스 변수
        self.gender = gender
        print('Person 생성자 호출(객체생성==클래스형 변수 생성==인스턴스 메모리생성)')
    def __del__(self):
        print('Person객체 소멸')
    def print_info(self):
        print(f"{self.name}님은{self.gender}입니다.")
    def __str__(self):
        return f"{self.name}은{self.gender}이죠."
````

````python
p1 = Person('신사임당','여자')
p1.print_info()
print(p1.__str__())
print(p1)
````

# 3절. 변수와 메소드
- 인스턴스 변수(맴버 변수) VS 클래스변수 VS static변수(정적변수)
- 인스턴스 메소드(self) VS 클래스매소드(cls) VS static메소드()

````python
# 클래스변수 : 클래스 내부 선언
# 클래스, 클래스변수를 수정하면 모든 인스턴스에 반영될수있음
# 객체변수, 클래스변수를 수정하면 해당 객체에만 반영
# static 변수 : 클래스 내부 선언.static변수 앞에 under score 2개(__)를 붙임

class Person:
    academy = 'MBC' #클래스 변수
    def __init__(self,name="신길동",gender="남자"): #self:내 객체자신
        self.name = name
        self.gender = gender
        print('Person 객체 생성합니다')
    def __str__(self):
        return("%s의 %s은(는) %s입니다." %(self.academy,self.name,self.gender))
print(Person)
print(Person.academy)
````

````python
p1=Person()
p2=Person()
````

````python
print(p1)
print(p2)
````
