````python
%%time
import time
start = time.time()
cnt = 0
while cnt<=10000000:
    cnt+=1
end = time.time()
print('반복문 1천만번 돈 시간:',(end-start),'초')
````

````python
del time #time 모듈을 메모리에서 unload
````
