```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.sample(2)
```

```python
plt.rcParams['figure.figsize']=(10,2)
plt.plot(iris.index,iris.iloc[:,:-1])
plt.legend(iris.columns[:-1],bbox_to_anchor=(1.05,1),loc=2,borderaxespad=0)
plt.show()
```

```python
g=iris.plot(kind='line',
           figsize=(12,3),
           color=['blue','pink','gold','aqua'],
           )
g.legend(loc='center left', bbox_to_anchor=(1,0.5))
plt.show()
```

```python
iris.plot(kind='box',
         colormap='winter',
         title='IRIS BOXplot',
         ylabel='y-value',
         xlabel='species',
         fontsize=15)
plt.show()
```

```python
iris.corr(numeric_only=True)
```

```python
iris.plot(kind='scatter',x='petal_length',y='petal_width',ylim=[-1,3])
plt.show()
```

```python
iris.plot(kind='scatter',x='petal_length',y='petal_width',
         s='sepal_length',
         c='sepal_length',
         ylim=[-1,3])
plt.show()
```

```python
iris.groupby('species').plot(kind='box',
                            subplots=True,
                            sharey=True,
                            )
plt.show()
```

```python
plt.figure(figsize=(12,4))
plt.scatter(iris.petal_length,iris.petal_width,
           s=iris.petal_length*30,
           c=iris.petal_width*20,
           alpha=.5)
plt.show()
```

```python
fig,axes = plt.subplots(ncols=2)
axes[0].boxplot(iris.sepal_width)
axes[0].violinplot(iris.sepal_width)
axes[1].violinplot(iris.sepal_width)
plt.show()
```

```python
x=[1,2,3]
y=[3,4,5]
fig,axes = plt.subplots(ncols=2,figsize=(12,3))
axes[0].bar(x,y,color='r')
axes[0].plot(x,y,c='k')
axes[1].barh(x,y,color='b')
axes[1].plot(x,y,'k')
```
