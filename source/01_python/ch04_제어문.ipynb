# 1절. 조건문(if)
```
if 조건1:
   조건1이 참일 경우 수행할 명령어들(if문 단독 가능, 들여쓰기 중요)
elif 조건2:
    조건1이 거짓이고, 조건2가 참일 경우 수행할 명령어들(elif는 필요한 만큼 사용가능) 
else :
    모든 조건이 만족하지 않을때 수행할 명령어들(else는 단독사용불가)
```

## 2.2 while 문
```
while 조건:
    조건이 참일 경우 반복할 문장들
else : 
    조건이 거짓일 경우 한번 실행할 문장(else선택사항)
```    

````python
# 1~100 까지 짝수들의 합을 게산하고 출력하시오(while문)
tot = 0 #누적변수
i=1
while i<=100:
    if i%2==0:
        tot +=i
    i+= 1 #i 변수 1증가    
else :
     print('1~100까지 짝수의 합은',tot)
````

````python
# ● 사용자에게 두 정수를 입력받아 작은수부터 큰수까지 홀수들의 합을 누적하여 출력하시오
first = int(input('첫번째 정수는?'))
second = int(input('두번째 정수는?'))
# (1) 두 수 중 작은 수는 minNo, 큰수는 maxNo에 할당
if first<second:
    minNo = first
    maxNo = second
else :
    minNo = second
    maxNo = first
# minNo = first if first<second else second
# maxNo = first if first>second else second
#(2) minNo~maxNo까지 홀수 누적하기  
tot = 0
for no in range(minNo,maxNo+1):
    if no%2: # if no%2!=0
        tot +=no
print(f"{minNo}부터{maxNo}까지 홀수의 합은 {tot}입니다.")        
````

````python
# ex-1~100사이의 정수를 맞추는 프로그램을 구현하시오(break)
'''
(프로그램 안에서 random수가 만들어지고 그 random수 맞추기)
1단게 : 수를 못 맞추면 UP/DOWN을 출력하고, 맞추면 축하 메세지 뿌리고 끝남(1~100사이의 수를 알아맞추기)
2단계 : 수를 못 맞추면 min~max를 조정, 맞추면 축하메시지 뿌리고 끝남
3단계 : 2단게의 기회르르 5회로 제한. 5회 기회가 끝나면 실패 메시지 뿌리고 끝남
(2단계와 3단계는 연습문제로)
'''
import random
randomNum = random.randint(1,100)
print('컴퓨터가 발생한 임의의 수:',randomNum)
while True :
    inputNum = int(input("1~100사이의 수를 맞춰보세요:"))
    if (inputNum>100) | (inputNum<1):
        print('수의 범위를 벗어났습니다.')
    if inputNum>randomNum:
        print("DOWN↓")
    elif inputNum<randomNum:
        print("up↑")
    else:
        print('축하 축하. 맞추셨습니다.')
        break
print('DONE')        
````

````python
# ex-1~100사이의 정수를 맞추는 프로그램을 구현하시오(break)
'''
(프로그램 안에서 random수가 만들어지고 그 random수 맞추기)
1단게 : 수를 못 맞추면 UP/DOWN을 출력하고, 맞추면 축하 메세지 뿌리고 끝남(1~100사이의 수를 알아맞추기)
2단계 : 수를 못 맞추면 min~max를 조정, 맞추면 축하메시지 뿌리고 끝남
3단계 : 2단게의 기회르르 5회로 제한. 5회 기회가 끝나면 실패 메시지 뿌리고 끝남
(2단계와 3단계는 연습문제로)
'''
import random
randomNum = random.randint(1,100)
print('컴퓨터가 발생한 임의의 수:',randomNum)
inputNum = 0
while inputNum!=randomNum :
    inputNum = int(input("1~100사이의 수를 맞춰보세요:"))#2단계는 min~max사이의 수를 맞춰보세요
    if (inputNum>100) | (inputNum<1):
        print('수의 범위를 벗어났습니다.')
    if inputNum>randomNum:
        print("DOWN↓")
        #2단계는 max값 조정
    elif inputNum<randomNum:
        print("up↑")
        #2단계는 min값 조정
    else:
        print('축하 축하. 맞추셨습니다.')
print('DONE')        
````

````python
# ex. 1~100사이의 정수를 맞추는 프로그램을 구현하시오
'''
(프로그램 안에서 random수가 만들어지고 그 random수 맞추기)
1단계 : 수를 못 맞추면 UP/DOWN을 출력하고, 맞추면 축하메시지 뿌리고 끝남 (1~100 사이의 수를 알아맞추기)
2단계 : 수를 못 맞추면 min~max를 조정, 맞추면 축하메세지 뿌리고 끝남
3단계 : 2단계의 기회를 5회로 제한. 5회의 기회가 지나면 실패메세지 뿌리고 끝남
(2단계와 3단계는 연습문제로)
'''
import random
randomNum = random.randint(1, 100)
print('컴퓨터가 발생한 임의의 수 :', randomNum)
inputNum = 0
while inputNum!=randomNum:
    inputNum = int(input("1~100 사이의 수를 맞춰 보세요 :")) # 2단계는 min~max사이의 수를 맞춰보세요
    if (inputNum>100) | (inputNum<1):
        print('수의 범위를 벗어났습니다')
    elif inputNum>randomNum:
        print("DOWN ↓")
        # 2단계는 max값 조정
    elif inputNum<randomNum:
        print("UP ↑")
        # 2단계는 min값 조정
    else:
        print('축하 축하. 맞추셨습니다')
print('DONE')
````

````python
# ex. 1~100사이의 정수를 맞추는 프로그램을 구현하시오
'''
(프로그램 안에서 random수가 만들어지고 그 random수 맞추기)
1단계 : 수를 못 맞추면 UP/DOWN을 출력하고, 맞추면 축하메시지 뿌리고 끝남 (1~100 사이의 수를 알아맞추기)
2단계 : 수를 못 맞추면 min~max를 조정, 맞추면 축하메세지 뿌리고 끝남
3단계 : 2단계의 기회를 5회로 제한. 5회의 기회가 지나면 실패메세지 뿌리고 끝남
(2단계와 3단계는 연습문제로)
'''
import random
randomNum = random.randint(1, 100)
print('컴퓨터가 발생한 임의의 수 :', randomNum)
inputNum = 0
min_Num = 1
max_Num = 100
while inputNum!=randomNum:
    inputNum = int(input(f"{min_Num}~{max_Num}사이의 수를 맞춰 보세요 :")) # 2단계는 min~max사이의 수를 맞춰보세요
    if (inputNum>max_Num) | (inputNum<min_Num):
        print('수의 범위를 벗어났습니다')
    elif inputNum>randomNum:
        print("DOWN ↓")
        # 2단계는 max값 조정
        max_num = inputNum -1
    elif inputNum<randomNum:
        print("UP ↑")
        # 2단계는 min값 조정
        min_num = inputNum + 1
    else:
        print('축하 축하. 맞추셨습니다')
print('DONE')
````
