````python
#args앞의 * : 매개변수들(인수들)이 args튜플로 전달
def add(*args):
    "전달받은 매개변수들의 합을 return.매개변수가 없을경우 0을 return"
    #print(args)#()
    tot = 0 # 누적변수
    for num in args:
        tot +=num #tot = tot + num
    return tot
````

## 3.7 인수의 언패킹
### 튜플인수 언패킹
- 패킹 : 함수 정의시 매개변수 앞에 * 붙임(튜플매개변수)
- 언패킹 : 함수호출시 매개변수앞에 * 붙임(리스트나 튜플이 각자 요소별 매개변수로) 

````python
def add(*args):
    tot = 0 # 누적변수
    for num in args:
        tot += num
    return tot
````

````python
def map_template(func,L=[]):
    result = []
    for item in L:
        result.append(func(item))
    return result
````

### (2) filter

````python
def filter_template(func,L):
    result = []
    for item in L:
        if func(item):
            result.append(item)
    return result
````

````python
def even_chk(value): #짝수면 True return,홀수면 False를 return
    return value%2==0
even_chk = lambda value:value%2==0
even_chk(9)
````
