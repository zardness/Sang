## 5.1 산술연산자:+-*/ //(몫연산자),**(제곱),%(나머지)

````python
print('10**5(승)=',10**5)
print('4/2(결과실수)=',4/2)
print('4/2(몫)=',4%2)
print('4%2(나머지)=',4%2)
````

````python
# 나누기를 제외한 산술연산은 정수끼리 결과는 정수, 정수와 실수 연산결과는 실수
print('10**5.0=',10**5.0)
print('4*2=',4*2)
print('4*2.0=',4*2.0)
````

````python
#quiz1.정수를 입력받아 찍수인지 홀수인지 여부를 출력
num = int(input('정수를 입력하세요>'))
result = num%2
if num%2:
    temp = '홀수'
else:
    temp = '짝수'
print("입력하신 수는 {}, 2로나눈 나머지는 {}, 그래서 {}를 입력하셨군요".format(num,result,temp))
print(f"입력하신 수는 {num}, 2로나눈 나머지는 {result}, 그래서 {temp}를 입력하셨군요")
````

````python
#quiz2.정수를 입력받아 5의배수 인지 여부를 출력
num = int(input('정수를 입력하세요>'))
result = num%5
if num%5:
    temp = '5의 배수가 아닙니다.'
else:
    temp = '5의 배수입니다.'
print("입력하신 수는 {}, 5로나눈 나머지는 {}, 그래서 {}를 입력하셨군요".format(num,result,temp))
print(f"입력하신 수는 {num}, 5로나눈 나머지는 {result}, 그래서 {temp}를 입력하셨군요")
````

````python
#quiz2.정수를 입력받아 5의배수 인지 여부를 출력
num = input('정수를 입력하세요')
if num[-1]=='0'or num[-1]=='5':
    print('5의 배수 입력')
else:
    print('5의 배수 아님')
````

````python
#quiz3.국,영,수 점수를 입력받아 다음셀에서 다음과 같이 출력하도록 구현하시오
#국어:99    영어:100   수학:99
#총점:299   평균:99.3
kor = int(input('국어?'))
eng = int(input('영어?'))
mat = int(input('수학'))
total = kor + eng + mat
avg = total /3
````

````python
print("국어:{:3d}\t영어:{:3}\t수학{:3}".format(kor, eng, mat))
print(f"총점:{total:3}\t평균:{avg:5.1f}")
````

## 5.3 논리연산자:and,&,or,|,not,

````python
# &  : 논리연산자
print((10>3) & (10>5)) # 논리연산 True & True =>True
print( 3 & 10) #비트연 수&수
#  3: 0011
# 10: 1010
# & : 0010(2)
print( 10>3 & 10>5) # 10>2>5
````

````python
# and : 논리연산자, 거짓으로 판별되는 첫번째 항의 결과 반환
#       모든항이 참이면 마지막 항을 반환
print(True and True and 'abc')
print(0 and 5)
````

````python
# |  : 논리연산자
print((10>3) | (10>5)) # 논리연산 True & True =>True
print( 3 | 10) #비트연 수&수
#  3: 0011
# 10: 1010
# | : 1011(11)
print( 10>3 | 10>5) # 10>11>5
````

````python
# or : 논리연산자, 참으로 판별되는 첫번째 항의 결과 반환
#       모든항이 거짓이면 마지막 항을 반환
print(False or False or 0)
print('홍길동')
````
