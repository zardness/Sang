## 5.3 논리연산자:and,&,or,|,not,

````python
# &  : 논리연산자
print((10>3) & (10>5)) # 논리연산 True & True =>True
print( 3 & 10) #비트연 수&수
#  3: 0011
# 10: 1010
# & : 0010(2)
print( 10>3 & 10>5) # 10>2>5
````

````python
# and : 논리연산자, 거짓으로 판별되는 첫번째 항의 결과 반환
#       모든항이 참이면 마지막 항을 반환
print(True and True and 'abc')
print(0 and 5)
````

````python
# |  : 논리연산자
print((10>3) | (10>5)) # 논리연산 True & True =>True
print( 3 | 10) #비트연 수&수
#  3: 0011
# 10: 1010
# | : 1011(11)
print( 10>3 | 10>5) # 10>11>5
````

````python
# or : 논리연산자, 참으로 판별되는 첫번째 항의 결과 반환
#       모든항이 거짓이면 마지막 항을 반환
print(False or False or 0)
print('홍길동')
````

# 6절 문자열 다루기
- 문자 슬라이싱 변수[from:stop:step]
- len(문자변수) : 문자수 
- upper() : 대문자로
- lower() : 소문자로
- title() : 각 어절의 앞글자만 대문자로
- capitalize() : 첫문짜만 대문자로
- count('찾을 문자열') : '찾을 문자열이,몇개 들어있는지 갯수
- find("찾을 문자열') : '찾을문자열'이 왼쪽부터 시작하여 몇번째에 있는지 index. 없으면-1
- rfind('찾을문자열') : '찾을문자열'이 오른쪽부터 시작하여 몇번째에 있는지 index. 없으면-1
-    cf. index('찾을문자열') : '찾을문자열'이 오른쪽부터 시작하여 몇번째에 있는지 index. 없으면 error
- startswith('특정문자열') : '특정문자열'로 시작하는지 여부
- endswith('특정문자열') : '특정문자열'로 끝나는지 여부

- strip() : 좌우공백제거(trim)
- rstrip() : 오른쪽 공백 제거
- lstrip() : 왼쪽 공백 제거
- split(나눌문자) : 공백을 기준으로 나누어 리스트로 반환
- isdigit() : 문자열이 숫자 문자열인지여부
- islower() : 문자열이 다 소문자인지 여부
- isupper() : 문자열이 다 대문자인지 여부
- replace(oldstr, newstr) : 문자열안의 oldstr을 newstr로 변경한 값을 반환
