```python
def safe_index(lst,item):
    '''
    첫번째 매개변수 lst에서 item 요소가 있는 index를 변환하는 함수
    item 요소가 없으면 -1을 반환
    '''
    if item in lst:
        return lst.index(item)
    return-1
```

```python
def safe_remove(lst, remove_item):
    '''
    첫번째 매개변수 lst에서 remove_item 데이터 제거
    해당 remove_item이 lst에 없으면 메세지 출력
    '''
    if remove_item in lst:
        lst.remove(remove_item)
    else:
        print('해당 데이터는 없는 데이터입니다.')
```

## 3.2 딕셔너리 item 수정,추가,제거

```python
my_favoite = {'fruit':'appla','number':12,'sport':'yoga'}

# 한 item씩 추가 
my_favoite['people']='홍길동'
print('1.한 item 추가후:',my_favoite)

# 여러 아이템을 한꺼번에 추가
my_favoite.update({'subject':'수학','major':'AI'})
print('2.여러 item 추가후:',my_favoite)

# 한item씩 수정 
my_favoite['fruit']='banana'
print('3.한item 수정후:',my_favoite)

# 여러 item을 한꺼번에 수정
my_favoite.update({'fruit':'kiwi','number':'-99'})
print('4.여러 item 수정후:',my_favoite)

# item 제거 방법1 : del변수 ['삭제할key']-없는 key일 경우 에러
del my_favoite['fruit']
if 'fruit'in my_favoite:
    del my_favoite['fruit']
else:
    print('5.fruit삭제하여 존재하지않음')

# item 제거방법2 : 변수.pop('삭제할key') - 해당 key값을 반환하고 제거(없는 key일경우 에러)
print('6.제거한 데이터:',my_favoite.pop('number'))
if'number' in my_favoite:
    print('7.제거한 데이터:',my_favoite.pop('number'))
else:
    print('7.number 삭제하여 존재하지 않음')
print(my_favoite)    
```

## 3.3 출력하기
- for문 이용

```python
my_tuple=('홍길동',30,'01099999999')# {0:'홍길동',1:30,2:01099999999}
for data in my_tuple:
    print(data,end='\t')
print() #개행 
for idx in range(len(my_tuple)):#[0,1,2]
    print("{}번째 : {}\t".format(idx,my_tuple[idx]),end="")
print() #개행 
for idx, value in enumerate(my_tuple):
    print(f"{idx}번째:{value}\t",end="")
```

## 4.4 셋 연산자:합집합(|), 교집합(&),차집합(-)

```python
s1={1,2,3,4,5}
s2={3,4,5,6,7}
print('s1=',s1)
print('s2=',s2)
print('s1 u s2=',s1|s2)# |,& : 집합연산(연산자앞뒤에 셋), 비트연산(연산자앞뒤에 숫자),논리연산(연산자앞뒤에 bool)
print('s1 & s2=',s1&s2)
print('s1-s2=',s1-s2)
```

```python
for idx,(name,tel,add)in enumerate(zip(names,tels,adds)):
    print(f"{idx+1}.{name}의 연락처는{tel}이고,{add}에 살아요")
```

```python
# names, tels, adds를 딕셔너리 리스트로 
members = []
for name, tel, add in zip(names,tels,adds):
    member ={'name':name, 'tel':tel,'address':add}
    members.append(member)
members    
```
