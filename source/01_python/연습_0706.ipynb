```python
Person1 = {'name':'홍길동','gender':'남자','academy':'MBC'}
Person2 = {'name':'홍길동','gender':'여자','academy':'MBC'}
print(Person1.get('name'),Person1['gender'],Person1.get('academy','-'))
print(Person1)
print(Person2)
```

```python
class Person:
    academy = "MBC" #클래스 변수
    def __init__(self,name,gender): #self:내 객체자신
        self.name = name
        self.gender = gender
        print('Person 객체 생성합니다.')
    def print_info(self):
        print("{}의 {}은/는 {}입니다.". format(self.academy,
                                        self.name,
                                        self.gender))
```

```python
p1 = Person('홍길동','남자')
print(p1)
p1.print_info()
p1.name='김길동'
p1.academy ='다른곳'
p1.print_info()
```

```python
p2=Person('김길순','여자')
p2.print_info()
p2.name="김길쑥"
p2.print_info()
```

```python
p1.academy,p2.academy,Person.academy
```

```python
class Person:
    def __init__(self,name,gender):
        self.name = name
        self.gender = gender
        print('Person 생성자 호출(객체생성==클래스형 변수 생성==인스턴스 메모리생성)')
    def __del__(self):
        print('Person객체소멸')
    def print_info(self):
        print(f"{self.name}님은{self.gender}입니다.")
    def __str__(self):
        return f"{self.name}은{self.gender}이죠" 
```

```python
p1 = Person('신사임당','여자')
p1.print_info()
print(p1.__str__())
print(p1)
```

```python
type(p1),isinstance(p1,Person)
```

```python
class Person:
    def __init__(self,name='아무개',gender='남자'):
        self.name = name
        self.gender="남자"
    def print_info(self):
        print(f"{self.name}님은 {self.gender}입니다.")
    def __str__(self):
        return f"{self.name}은{self.gender}이네요"
```
